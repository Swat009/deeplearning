```python
vocab_size=5000
```

```python
trn=[np.array([i if i<vocab_size-1 else vocab_size-1 for i in s]) for s in x_train]
test=[np.array([i if i<vocab_size-1 else vocab_size-1 for i in s]) for s in x_test]
```

```python
lens=np.array(list(map(len,trn)))
```

```python
lens
```

```python
(lens.max(),lens.min(),lens.mean())
```

```python
seq_len=500
```

```python
trn
```

```python
trn=sequence.pad_sequences(trn,maxlen=seq_len,value=0)
test=sequence.pad_sequences(test,maxlen=seq_len,value=0)
```

```python
#trn[0]
```

```python
##Simple Model
```

```python
model=Sequential([Embedding(vocab_size,32,input_length=seq_len),Flatten(),Dense(100,activation='relu'),Dropout(0,7),Dense(1,activation='sigmoid')])
```

```python
model.compile(loss='binary_crossentropy',optimizer=Adam(),metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(trn,labels_train,validation_data=(test,labels_test),nb_epoch=1,batch_size=10)
```

```python
## Convolutional neural network
```

```python
conv1=Sequential([Embedding(vocab_size,32,input_length=seq_len,dropout=0.2),
                  Dropout(0.2),
                  Convolution1D(64,5,border_mode='same',activation='relu'),
                  Dropout(0.2),
                  MaxPooling1D(),
                  Flatten(),
                  Dense(100,activation='relu'),
                  Dropout(0.7),
                  Dense(1,activation='sigmoid')
                 ])
```

```python
conv1.summary()
```

```python
conv1.compile(loss='binary_crossentropy',optimizer=Adam(),metrics=['accuracy'])
```

```python
conv1.fit(trn,labels_train,validation_data=(test,labels_test),nb_epoch=1,batch_size=10)
```
